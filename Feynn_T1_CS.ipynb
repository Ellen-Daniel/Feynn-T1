```python
plt.figure(figsize=(12, 6))
plt.bar(df['State/ UTs'], df['No. of charging stations'])
plt.xticks(rotation=90)
plt.xlabel('State/UTs')
plt.ylabel('No. of Charging Stations')
plt.title('Distribution of Charging Stations across States/UTs')
plt.show()
```

```python
total_per_state = df.groupby('State/ UTs')['No. of charging stations'].sum()
average_charging_stations = df['No. of charging stations'].mean()
max_charging_stations = df.loc[df['No. of charging stations'].idxmax()]
min_charging_stations = df.loc[df['No. of charging stations'].idxmin()]
```

```python
print("Total number of charging stations per State/ UTs:\n", total_per_state)
print("\nAverage number of charging stations:", average_charging_stations)
print("\nState/UTs with the maximum number of charging stations:\n", max_charging_stations)
print("\nState/UTs with the minimum number of charging stations:\n", min_charging_stations)
```
